## Manual cleaning process in more detail:
Load both the silver standard (non-IA adjudicated) main project and gold standard adjudicated IA project.

Clean the main project from empty and redundant docs to obtain the final corpus.
Joins the final docs into a corpus dict

 `{"gold": list(Documents), "silver": list(Documents)}.`
 
 Gold-standard: adjudicated from IA study based off three annotators.
 Silver-standard: annotated by 1 annotator and corrected by judge.
Corpus creation procedure is as follows:
1. Set aside gold standard IAA documents.
2. Remove IA docs from silver standard.
3. Remove documents without annotations: these are an artifact of WebAnno monitoring.

````python
from parser import WebannoProject
from sentivent_webannoparser import util
from sentivent_webannoparser import parse_project
import settings
from itertools import groupby

def check_dupes(all_docs):
    # title_key = lambda x: x.title.split("_")[-1].replace("-", "")[:32]
    title_key = lambda x: x.path.split("/")[-3].replace("-", "")[:32]
    all_docs.sort(key = title_key)
    i = 0
    j = 0
    for title, g in groupby(all_docs, title_key):
        j += 1
        g = list(g)
        if len(g) > 1:
            i += 1
            print(f"------------------------------------------------------\n{i}. {title}\n------------------------------------------------------")
            dupes = []
            for d in g:
                dupes.append((d.title, d.annotator_id, str(len(d.events)) if d.events else "0", d.path))
            dupes.sort(key=lambda x: x[2], reverse=True)
            print("\n".join("- " + " ".join(d) for d in dupes))
    print(f"{j} different source documents, {len(all_docs)} total annotated of which {len(all_docs)-j} are annotated by multiple annotators.")

def split_gold_silver(docs):
    '''
    Takes a list of docs and checks wether they are in gold-standard (from IAA adjudicated) or not.
    return: two lists gold_docs and silver_docs
    '''
    gold = []
    silver = []
    for doc in docs:
        if doc.title in settings.IA_IDS:
            if doc.annotator_id == settings.MOD_ID: # throw away any IAA study that is not gilles
                gold.append(doc)
        else:
            silver.append(doc)
    return gold, silver
print("Succesfully imported modules and functions.")
````

````python
from datetime import datetime
from shutil import copy2, copytree, rmtree
from pathlib import Path
import zipfile

base_export_dirp = settings.FINAL_XMI_DIRP # use final_project as base project to copy from

timestamp = datetime.today().strftime('%Y-%m-%d_%H%M%S')
final_clean_optdirp = Path().cwd() / Path(settings.OPT_DIRP) / f"XMI-SENTiVENT-event-english_{timestamp}"
final_doc_dir = final_clean_optdirp / "annotations_to_import/"

# copy dir structure and files except those in base/annotation/.
base_ann_fp = Path(final_clean_optdirp) / "annotation/"
if not final_clean_optdirp.exists():
    copytree(base_export_dirp, final_clean_optdirp)
rmtree(final_clean_optdirp / "annotation") # easier to remove the folder than using copytree's ignore func

# mkdir
final_doc_dir.mkdir(parents=True)

# now place files from all_docs in base/annotation/
src_fps = [doc.path for doc in all_docs_clean]
dest_fps = [base_ann_fp / "/".join(fp.split("/")[-3:-2]) for fp in src_fps]
tmp_dest_fps = [Path("/tmp/") / "/".join(fp.split("/")[-3:-2]) for fp in src_fps]
src_fps = [Path().cwd() / "/".join(fp.split("/")[:-1]) for fp in src_fps]
for src_fp, dest_fp, temp_fp in zip(src_fps, dest_fps, tmp_dest_fps):
    print(f"Copying {src_fp} > {dest_fp}.")
    dest_fp.mkdir(parents=True)
    copy2(src_fp, dest_fp)
# unzip and place in temp dir in order to properly name file
    print(f"Extract to {temp_fp}.")
    with zipfile.ZipFile(src_fp, 'r') as zip_ref:
        zip_ref.extractall(temp_fp)
# get file and rename and move from tmp
    unzip_fp = next(temp_fp.glob("**/*.xmi"))
    unzip_fn = str(unzip_fp.parts[2]) + "__" + str(unzip_fp.parts[3])
    unzip_dest_fp = final_doc_dir / unzip_fn
    print(f"Move to {unzip_dest_fp}.")
    copy2(unzip_fp, unzip_dest_fp) 
# clear tmp
    rmtree(temp_fp)
````
